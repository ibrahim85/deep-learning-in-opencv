```python
import numpy as np
import cv2
```

```python
net = cv2.dnn.readNetFromCaffe("../../pretrained-models/bvlc_googlenet.prototxt",
                               "../../pretrained-models/bvlc_googlenet.caffemodel")
```

```python
print("gflops: ", net.getFLOPS((1, 3, 224, 224)) * 1e-9)
```

```python
w, b = net.getMemoryConsumption((1, 3, 224, 224))
print("weights (mb):", w*1e-6, ", blobs (mb):", b*1e-6)
```

```python
blob = cv2.dnn.blobFromImage(np.zeros((224, 224, 3), np.uint8),
                                      1, (224, 224))
net.setInput(blob)
net.forward()
```

```python
# report total time
total, timings = net.getPerfProfile() # return total time and each layer's time
tick2ms = 1e3 / cv2.getTickFrequency()
print("inference (ms): {:2f}".format(total * tick2ms))
```

```python
# report inference time of every layer
layer_names = net.getLayerNames()
print("{: <30} {}".format("LAYER", "TIME (ms)"))
for (i, t) in enumerate(timings):
    print("{: <30} {:.2f}".format(layer_names[i], t[0] * tick2ms))
```

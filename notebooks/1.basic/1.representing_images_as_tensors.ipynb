The OpenCV Deep Learning functionality operates four-dimensional floating-point tensors with an NCHW order of dimensions:
- N for the number of images in the batch
- C for the number of channels
- H for the height
- W for the width

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image_bgr = cv2.imread("../../datasets/Buakaw.jpg")
image_rgb = image_bgr[:, :, [2, 1, 0]]

plt.figure(figsize=(6, 6))
plt.imshow(image_rgb)
plt.show()
```

```python
image_rgb.shape
```

```python
image_rgb = image_rgb.astype(np.float32)
tensor_chw = np.transpose(image_rgb, (2, 0, 1))
tensor_chw.shape
```

```python
tensor_nchw = tensor_chw[np.newaxis, ...]
tensor_nchw.shape
```
